```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

print("Bibliotecas Carregadas com Sucesso")
```

```python
arquivo='films.txt'
```

```python
df = pd.read_json(arquivo)
```

```python
df.head(50)
```

```python
df.dtypes
```

## 1.Crie uma visualização que mostre o orçamento de todos os filmes do conjunto de dados. Qual o filme com maior orçamento e com menor orçamento?

```python
df_orc = df.sort_values('Budget_M', ascending=False)
```

```python
df_orc.head()
```

```python
films = df_orc.Film
orc = df_orc.Budget_M
plt.figure(figsize=(13,8))

plt.title('Orçamentos por Filmes')
ax = sns.barplot(y=films, x=orc, data=df_orc, orient='h').set(xlabel="Orçamentos", ylabel="Filmes")
plt.show()
```

## Resposta:

#### Concluimos após a análise que: Pirates of the Caribbean: At World's End é o Filme com o Maior Orçamento e The Twilight Saga: New Moon é o Filme com Menor Orçamento.



## 2.Crie uma visualização que mostre a bilheteria de todos os filmes do conjunto de dados. Qual o filme com maior bilheteria e com menor bilheteria?

```python
df.head()
```

```python
df_bil = df.sort_values('Worldwide_Gross_M', ascending=False)
```

```python
films = df_bil.Film
bil = df_bil.Worldwide_Gross_M
plt.figure(figsize=(13,8))

plt.title('Bilheteria dos Filmes')
ax = sns.barplot(y=films, x=bil, data=df_bil, orient='h').set(xlabel="Bilheteria", ylabel="Filmes")
plt.show()
```

## Resposta:

#### Concluimos após a análise que: Avatar foi o Filme com maior bilheteria e Ratatouille foi o filme com menor bilheteria.
