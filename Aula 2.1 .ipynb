```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
categoria = ['A','B','C','D']
valores = [10, 20, 15, 5]
plt.bar(categoria, valores)
plt.xlabel('Grupos')
plt.ylabel('Tamanhos')
plt.title('Meu gráfico de Barras')
plt.show()
```

```python
plt.figure(figsize=(9, 3))
plt.bar(categoria, valores, color='darkred')
plt.xlabel('Grupos')
plt.ylabel('Tamanhos')
plt.title('Meu gráfico de Barras')
plt.show()
```

```python
filmes_uri = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
```

```python
filmes = pd.read_json(filmes_uri)
```

```python
filmes.head()
```

```python
filmes.describe()
```

```python
contagens = filmes.groupby(by='Genre').count()
```

```python
contagens
```

```python
plt.figure(figsize=(9, 5))
plt.bar(contagens.index, contagens['Worldwide_Gross_M'])
plt.show()
```

```python
# outros estilos possíveis são: ‘darkgrid’, ‘whitegrid’, ‘dark’, ‘white’ e ‘ticks
sns.set(style="darkgrid")
sns.countplot(x="Genre", data=filmes)
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10,5))
plt.title('Quantidade de filmes por gênero')
sns.countplot(x="Genre", color = 'r', data=filmes)
plt.show()
```

```python
plt.title('Quantidade de filmes por gênero')
sns.countplot(y="Genre", color = 'r', data=filmes)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Bilheteria média por gênero')
sns.barplot(x='Genre', y='Worldwide_Gross_M', color = 'b', ci = 'sd', data=filmes)
plt.show()
```

```python
# countplot() moostra uma contagem de elementos de grupos
# barplot() mostra medidas associadas a esse grupos
# por padrão a medida é MÉDIA ARITIMÉTICA
```

```python
plt.figure(figsize=(10,5))
plt.title('Bilheteria total por gênero')
sns.barplot(x="Genre", y="Worldwide_Gross_M", color = 'orange', estimator = sum, ci = None, data=filmes)
plt.show()

# estimator = sum , tira o padrão da média e apresenta como soma
```
